2. Fill in the missing entries. For filling any feature, you can use either the mean or median value of the feature values from observed entries. Explain the reason behind your choice and print replacement value of each feature.

I have calculated the mean of the column and filled the value in the missing place. The mean provides the average of the column so it is easy to keep the missing value in range according to the  other values mentioned in the column. 

```python
mean = data.loc[:, 'f21'].mean()
dataset = data.fillna(mean)
dataset.isna().any().sum() 
```
